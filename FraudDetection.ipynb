```python
#Class Level

#0 --> Normal Transation
#1 --> Fraudulent Transaction
normal=df[df.Class==0]
fraud=df[df.Class==1]
```

```python
#measure of normal transactions
normal.Amount.describe()
```

```python
#measure of fraud transactions
fraud.Amount.describe()
```

```python
df.groupby('Class')['Amount'].mean()
#mean of legit transaction is 88
#where fraud transaction mean is 122
#mean of normal transaction is very different from mean of fraudelent transaction
```

```python
# sample dataset from legit transaction using Random Sampling 
#and join it with fraudelent dataset to make uniform distribution
normalsample=normal.sample(n=492)
normalsample.shape
```

```python
data = pd.concat([normalsample, fraud], axis=0)
#axis=0 joins rows
```

```python
data.head(5)
```

```python
data.tail(5)
```

```python
data['Class'].value_counts()
```

```python
data.groupby('Class')['Amount'].mean()
```

```python
#we have got a good sample as sample mean value is similar to population mean value
#importing model and train_test_split

from sklearn.model_selection import train_test_split   #importing train_test_split for spliting data
from sklearn.linear_model import LogisticRegression    #importing LogisticRegression as the problem is classification problem
from sklearn.metrics import accuracy_score             #importing performance matrics accury_score to check model accuracy
```

```python
#creating two variable and storing X= all features, Y= Class

X = data.drop(columns='Class', axis=1)        #axis=1 represent column
Y = data['Class']
```

```python
X
```

```python
Y
```

```python
#we will need to create four variable to split the data
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
 X_train.shape
 Y_train.shape
 X_test.shape
 Y_test.shape
```

```python
model = LogisticRegression()
model.fit(X_train, Y_train)
```

```python
#Accurracy on Training data

X_train_predict = model.predict(X_train)
Training_data_accurary = accuracy_score(X_train_predict, Y_train)
Training_data_accurary
```

```python
#checking accuracy on Test/unseen date


X_test_prediction = model.predict(X_test)
Test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
Test_data_accuracy
```

```python
# Logistic regression achieved high accuracy (91% on training, 89.2% on test), demonstrating its effectiveness for fraud detection.
```
